```
import numpy as np                                              # 导入数学函数
from matplotlib import pyplot as plt                        # 导入作图函数

# 数据录入——请在这里修改或补充每日病例数，数据太多时用"\"表示换行
num = (119, 159, 202, 257, 342, 442, 600, 729, 875, 1091, 1333, 1474, \
                    2095, 2822, 3412, 4045, 4759, 5654, 6316, 7448, 8591, 9787, 11283, 12545)
days = len(num)

rate = [0.3]
for i in range(1, days):
    rate.append((num[i]-num[i-1])/num[i-1])

print(rate)

rate_5 = rate[0:5]
for i in range(5, days):
    summary = sum(rate[j] for j in range(i, i-5, -1))
    rate_5.append(summary/5)
print(rate_5)

x = range(days)
std = [ ]
std.append(rate_5[0])
for i in range(1, days):
    std.append(rate_5[i-1]/(1+rate_5[i-1]))
print(std)

plt.plot(x, rate,  color='gray', label='daily increase rate')    
plt.plot(x, rate_5,  color='b', label='5 days moving average')    
plt.plot(x, std, color='r', label='critical line')
plt.xlabel('Day')                                               # 打印横坐标标签
plt.ylabel('Rate')                                              # 打印纵坐标标签
plt.title('Daily Increasing Rate Curve')                # 打印图表名称
plt.legend(loc='best')                                       # 打印图例说明
plt.show()

```
